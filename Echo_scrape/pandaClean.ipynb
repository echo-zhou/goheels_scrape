```python
cols = df.columns.tolist()
# make a list consist of all the columns
```

```python
name_index = cols.index('name')
name_index

# find the index of 'name'
```

```python
cols.insert(0,cols.pop(name_index))
cols.insert(1,cols.pop(cols.index('number')))

# put name in the first place (index 0 in the list)
# put player number in the second place

cols
```

```python
df = df[cols]
```

```python
df = df.apply(pd.to_numeric, errors='ignore')

# convert 'number' from string to integer so we can sort by number
```

```python
df.sort_values(['number'],inplace=True,ascending=[1])
df
```

```python
df.index = range(len(df))
df

#change the index from names to numbers
#range, a python function 
#len = length
```

```python
##############################
#add 'AvgPTS'(average points, calculated from personal STATs)
```

```python
df['AvgPTS'] = 1.1
# add a column called Avgpts, give it a value 1.1, so set the datatype to floating numbers
```

```python
for i in range(len(df)):
    if df.iloc[i,1]!=30:
        cur_sf = pd.read_csv("current_stats_of_{}.csv".format(df.iloc[i,1]))
        cur_avg = cur_sf['PTS'].mean()
    else:
        cur_avg = np.NaN
    df.iat[i,len(df.columns)-1] = cur_avg
    
# loc = location 
# iloc, a panda language
# a special player who's id is 30 doesn't have any stats, so need to loop him out
# if this person't id isn't 30, open his csv file with panda
# get column [PTS] in the CSV file, use function 'mean' to calculate the avg
# iat basically same as iloc, panda language
```
